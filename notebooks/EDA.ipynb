```python
import requests

# 사진이 포함된(photos=true) 한국(6891)의 사슴벌레(49083) 관찰 기록 조회
url = "https://api.inaturalist.org/v1/observations?taxon_id=49083&place_id=6891&photos=true&per_page=1"
response = requests.get(url).json()

print(f"[!] 대한민국 사슴벌레 전체 데이터 수: {response['total_results']}건")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import os

# 1. 데이터 로드
base_path = Path("..")
metadata_path = base_path / "data" / "raw" / "inaturalist" / "metadata.csv"
df = pd.read_csv(metadata_path)

# 2. 종별 데이터 개수 집계
class_counts = df['scientific_name'].value_counts()
print("--- Class Distribution ---")
print(class_counts)

# 3. 시각화 (현업 보고용 스타일)
plt.figure(figsize=(12, 8))
sns.barplot(x=class_counts.values, y=class_counts.index, palette='viridis')
plt.title('Stag Beetle Species Distribution (iNaturalist)', fontsize=15)
plt.xlabel('Number of Images', fontsize=12)
plt.ylabel('Species Name', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
